```python
from  sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = 10000

x, y = make_classification(
    n_samples = points,
    n_features = 2,
    n_informative = 1,
    n_redundant = 0,
    n_classes = 2,
    n_clusters_per_class = 1,
    random_state = 0,
    hypercube = False,
    class_sep = 10,
)
plt.figure(figsize = (8, 6))
plt.scatter(x[:,0], x[:,1], c = y,  cmap = 'viridis', s = 8)
```

```python
def step(x):
    return 1 if x > 0 else 0

def perceptron(x, y):
    x = np.insert(x, 0, 1, axis = 1) # inserting the bias or 1 weight to make the dimension similar to line representation which is  w0 or intercept = 1, w1 = x[0], w2 = x[1]
    weights = np.ones(x.shape[1])  # at the start our weights are [1, 1, 1]
    learning_rate = 0.1
    for i in range(10000):
        j = np.random.randint(0,points) # choosing a random element/point
        y_cap = step(np.dot(x[j], weights))
        weights = weights + learning_rate * ( y[j] - y_cap) * x[j]

    return weights[0], weights[1:]
```

```python
# getting the slope and plotting the resultant line
'''
let's say our line  is ax + by + c = 0 , then (a, b) are coefficients and c is intercept,
slope becoms m = -a/b
'''
intercept, coefficient = perceptron(x, y)
a, b = coefficient
m = -a/b
c = -intercept/b

# line now is y = m*x + c
x_input = np.linspace(-6, 8, points)  # making 1000 points
y_input = m * x_input + c

print(coefficient, intercept)

# plotting the line
plt.plot(x_input, y_input, color='orange', linewidth = 2)
plt.scatter(x[:,0], x[:,1], c = y, cmap='viridis', s = 10)
plt.show()
```
