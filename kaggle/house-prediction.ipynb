```python
test_url = '/kaggle/input/house-prices-advanced-regression-techniques/test.csv' 
train_url = '/kaggle/input/house-prices-advanced-regression-techniques/train.csv' 
house_train = pd.read_csv(train_url) 
house_test = pd.read_csv(test_url) 
titles = house_train.columns.tolist() 
length, dims = house_test.shape
dims
```

```python
# Dealing with missing values 
missing_columns = []
missing = house_train.isnull()   # getting all the missing value information 
for item in titles : 
    miss = missing[item].value_counts().to_dict()
    if miss[False] < length : missing_columns.append(item) # getting list of columns with missing values 

# missing column information 
# house_train[missing_columns].head(15) 
for col in missing_columns : 
    print(house_train[col].isnull().value_counts(), '\n\n\n' , house_train[col].value_counts(), end = '\n\n\n')

len(missing_columns)
```

```python
house_train[missing_columns].head(7)
```

### Grouping Various missing columns on the actions to be taken 

```python
cols_to_drop = ['Alley', 'MiscFeature','Fence','PoolQC'] 
cols_mode = [
    'FireplaceQu', 'MasVnrType', 'BsmtQual', 'BsmtCond', 
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
    'GarageQual', 'GarageCond', 'GarageType', 'GarageFinish'
]
cols_median = [
    'MasVnrArea','GarageYrBlt','LotFrontage'
]
len(cols_mode), len(cols_median), len(cols_to_drop), len(missing_columns)
```

### Dropping the columns 

```python
house_data = house_train.drop(cols_to_drop, axis = 1) 
print(house_data.shape)
house_train[cols_to_drop].head(15)
```

### Mode imputation of the columns 

```python
# Seeing modes of all the columns which are to be mode-imputed

# plotting all the cols to see the distribution 
for col in cols_mode:
    print(house_data[col].mode())
#     plt.figure(figsize=(4, 8))  # Adjust the figure size as needed
#     sns.histplot(data=house_data, x=col, kde=True)
#     plt.title(f'Distribution of {col}')
#     plt.xlabel(col)
#     plt.ylabel('Frequency')
#     plt.show()
```

### Imputing the values 

```python
for col in cols_mode: 
    mode_value = house_data[col].mode()[0]
    house_data[col].fillna(mode_value, inplace = True) 
    
# checking the no. of null values after imputation
for col in cols_mode: 
    print(house_data[col].isnull().value_counts())
```

### Median Imputation

```python
# since we can't visualize median using histograms or charts,
# for median visualization we can use box plots 
# for col in cols_median:
#     plt.figure(figsize = (8,6)) 
#     sns.histplot(data = house_data, x = col, kde = True) 
#     plt.show()

for col in cols_median: 
    plt.figure(figsize = (6,6)) 
    sns.boxplot(data = house_data, x = col) 
    median_val = house_data[col].median()
    print('Median : ', median_val)
#     plt.axhline(
#         y = median_val, 
#         color = 'maroon', 
#         linestyle = '--',
#         label = f'Median : {median_val}'
#     )
#     plt.legend()
    plt.show()
    
```

```python
# imputing the null values with their respective median values
for col in cols_median: 
    med_val = house_data[col].median()
    house_data[col].fillna(med_val, inplace = True) 


for col in cols_median:
    print(house_data[col].median())
    print(house_data[col].isnull().value_counts(), end = '\n\n\n')
```

```python
titles = house_data.columns.tolist()
for col in titles: 
    miss = house_data[col].isnull().value_counts()
    if 'True' in miss : print(col)

# since we don't have any column left with null values we are okay 
```
