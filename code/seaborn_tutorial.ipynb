```python
#@title Imports
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
sns.set_theme(style = 'darkgrid') # default, darkgrid, whitegrid, dark, white, ticks
```

```python
#@title Tips dataset
tips_data = sns.load_dataset('tips') # this is inbuilt data set in seaborn library
# we can set theme for the plots
# visualize the data
''' This will show smokers at the time of lunch/ dinner  for a tip/total_bill intersection.
it means how many people are smokers at the lunch time, whose total bill was this much and they paid this much amount of tip.'''
sns.relplot(data = tips_data, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size') # basically this is relational plot as the name suggests
```

```python
#@title Iris Dataset
iris_data = sns.load_dataset('iris')
# iris_data.head(10)
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data = iris_data)
sns.countplot(x='sepal_length',  data = iris_data)
# sns.histplot(x='sepal_length', data = iris_data, bins =20,  kde = False)
```

```python
#@title California House Dataset
''' loading the housing dataset of california housing from sklearn'''

from sklearn.datasets import  fetch_california_housing
house_california = fetch_california_housing()
house_data = pd.DataFrame(house_california.data, columns = house_california.feature_names)
house_data['PRICE'] = house_california.target
house_data
```

```python
#@title Boston Housing Dataset
data_url = 'https://raw.githubusercontent.com/krahul2024/machine-learning/main/data/HousingData.csv'
housing_data = pd.read_csv(data_url)
housing_data = housing_data.rename(columns = {'MEDV':'PRICE'})
housing_data.head()
```

Distribution Plot

```python
sns.displot(housing_data['PRICE']) # this shows distribution of prices  with the
sns.histplot(housing_data['PRICE']) # this does the same thing as displot
```

Correlation :
*   Positive correlation : Directional proportionality
*   Negative correltaion : Inverse proportionality




Heat map for this data

```python
''' This df.corr() gives us the correlation matrix for a dataframe and we then can plot
 the correlation matrix using matplot and see how are various aspects of the dataset related to each other'''
corr_matrix = housing_data.corr()
# print(corr_matrix)

# plotting the heat map
plt.figure(figsize = (16, 16))  # creating a matplot figure of size 8x8 for plotting the heatmap or correlation matrix
''' cbar - is color scale which shows values on color basis
    square - shape of each of the heatmap
    fmt - format of displaying of correlation values just like printf
    annot - displaying corresponding correlation values for each of the heatmap
    annot_kws - keywords for annotation like font-size
    cmap - for various sahdes of a color various options are Blues, viridis, cividis, coolwarm, crest '''
sns.heatmap(corr_matrix, cbar = True, square = True , fmt = '.3f', annot = True, annot_kws = {'size' :12}, cmap  = 'viridis')
```
